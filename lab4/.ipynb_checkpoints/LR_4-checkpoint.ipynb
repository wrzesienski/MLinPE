```python
# Задаем указатели для работы с временной осью по точкам
train_pointer = int(0.6 * AEW_data.shape[0]) # обучаемся на 60% от временной последовательности
val_pointer = train_pointer + 1 + int(0.2 * AEW_data.shape[0])
test_pointer = val_pointer + 1 + int(0.2 * AEW_data.shape[0])
```

<b>Нормализация данных: считаем среднее и делим на стандартное отклонение</b>

```python
mean = AEW_data[:train_pointer].mean(axis=0) # считаем среднее
AEW_data = AEW_data - mean
std = AEW_data[:train_pointer].std(axis=0) # Возвращает стандартное отклонение выборки по запрошенной оси
AEW_data = AEW_data/std
```

<h3>3. Написание функции-генератора для формирования окон из временного ряда данных.</h3>

```python
def generator(data, lookback, delay, min_index, max_index,
             shuffle=False, batch_size=128, step=1):
    
    if max_index is None:
        max_index = len(data) - delay - 1 # подстраховка от выхода за границу окна при предсказании вперед
    i = min_index + lookback # подстраховка от выхода за границу окна, сдвиг вперед + используется для предсказания
    
    while 1:
        
        if shuffle: # перемешивание
            rows = np.random.randint(
                min_index + lookback, max_index, size=batch_size) # в массиве будет size элементов
        else:
            # Если превысили макс. индекс строк
            if i + batch_size >= max_index: 
                i = min_index + lookback # перезапуск генератора - переходим в начало последовательности
            rows = np.arange(i, min(i + batch_size, max_index)) # чтобы не выйти за предел последовательности
            
        i += len(rows) # используется для следующего вызова генератора. Генератор запоминает i и при вызове генератора в след.раз 
            
        samples = np.zeros((len(rows),
                           lookback // step,
                           data.shape[-1])) #массивы размера тех точек, которые используются для обучения
        targets = np.zeros((len(rows),)) # массив ответов по samples
        
        #print(rows)
        
        # enumerate генерирует кортеж - (индекс элемента, элемент)
        for j, row in enumerate(rows): # здесь j - номер, row - элемент в последовательности от (i до i+batch_size) (либо до конца послед.)
            indices = range(rows[j] - lookback, rows[j], step) #
            samples[j] = data[indices]
            targets[j] = data[rows[j] + delay]
        
        yield samples, targets
```

<h3>4. Инициализация генераторов для задач обучения, валидации и тестирования.</h3>

```python
lookback = 144 
step = 2
delay = 8 
batch_size = 48 # ширина временного окна

#lookback = 240 
#step = 1
#delay = 6 
#batch_size = 24 

train_gen = generator(AEW, 
                      lookback=lookback, 
                      delay=delay, 
                      min_index=0,
                      max_index=train_pointer, 
                      shuffle=True,                      
                      step=step,                      
                      batch_size=batch_size) # случайные точки из обучающей области данных


val_gen = generator(AEW, 
                    lookback=lookback,  
                    delay=delay, 
                    min_index=train_pointer + 1,    
                    max_index=val_pointer,                    
                    step=step,                    
                    batch_size=batch_size)


test_gen = generator(AEW,                     
                     lookback=lookback,                   
                     delay=delay, 
                     min_index = val_pointer + 1,
                     max_index = None,    
                     step=step,                
                     batch_size=batch_size)
```

<b>Напишем функции, облегчающие выполнения повторяемых в работе действий</b>

```python
# функция для построения двух графиков, по предсказанной и реальной выборкам
def plotComparativeGraph(test_predict, test_AEW, length = 720 ): 

    pointer = val_pointer + lookback + delay + 1
    plt.figure(figsize=(10,6))

    plt.plot([i for i in range(length)], test_predict[:length])
    plt.plot([i for i in range(length)], test_AEW[:length])

    plt.title("Сравнительный график за " +str(length)+" часов (" + str(length//24)+" полных суток)")
    plt.xlabel('Часы')
    plt.ylabel('Потребление')
    plt.grid()
    plt.legend(['Спрогнозированный график', 'Реальный график'], loc = 'upper left')
    plt.show() 
```

```python
# средняя абсолютная ошибка

def getMae(predicted_AEW, test_AEW):
    batch_maes = []

    for i in range(len(test_AEW)):
        #mae = np.mean(np.abs(predicted_AEW[i] - test_AEW[i]))
        mae = np.abs(predicted_AEW[i] - test_AEW[i])
        batch_maes.append(mae)
    return np.mean(batch_maes)
```

```python
# средняя ошибка

def getMe(predicted_AEW, test_AEW):
    batch_maes = []

    for i in range(len(test_AEW)):
        mae = predicted_AEW[i] - test_AEW[i]
        batch_maes.append(mae)
    return np.mean(batch_maes)
```

```python
train_steps = (train_pointer + 1 - lookback) // batch_size # Сколько раз нужно обратиться к val_gen, #деление без остатка
#чтобы получить проверочный набор данных целиком
print(train_steps)
```

```python
val_steps = (val_pointer - (train_pointer + 1) - lookback) // batch_size # Сколько раз нужно обратиться к val_gen, 
#чтобы получить проверочный набор данных целиком
print(val_steps)
```

```python
test_steps = (len(AEW) - (val_pointer + 1) - lookback) // batch_size  #Сколько раз нужно обратиться к test_gen, 
#чтобы получить тестовый набор данных целиком
print(test_steps)
```

<h3>5. Проверка качества базового решения задачи без применения машинного обучения.</h3> <br>
<b>В данном случае предполагается, что следующее значение временного ряда равно фактическому значению предыдущего шага</b>

<b>Расчёт средней и абсолютной средней ошибки для решения задачи без применения машинного обучения</b>

```python
def mistakeByMae():
    batch_maes = []
    batch_mes = []
    predicted_values = []
    target_values = []
    for step in range(test_steps):
        samples, targets = next(test_gen)
        
        target_values.append(targets)
        
        preds = samples[:,-1, 0]  # последние элементы каждой выборки
        
        predicted_values.append(preds)
        
        mae = np.mean(np.abs(preds - targets))
        batch_maes.append(mae)
        
        me = np.mean(preds - targets)
        batch_mes.append(me)
                              
    return batch_maes, batch_mes, predicted_values, target_values

maes, mes, predicts, targets = mistakeByMae()
```

<h3>6. Обучение сети многослойного перцептрона прогнозирования временного ряда.</h3>

<p>Задание архитектуры сети. Для многослойного перцептрона необходима форма тензора входных данных в виде (размер пакета, признаки). Однако функция-генератор возвращает окна формы (размер_пакета, размер_окна, признаки), где признаками являются количество одновременно анализируемых временных рядов. Поэтому в качестве входного используется слой уплощения</p>

```python
print("Интервал в прошлом от заданного момента:", lookback)
print("Размер шага:", step)
print("Размерность пакета с признаками:",lookback // step)
print("Размерность пакета с ответами:",AEW.shape[-1])
```

<h3>10. Обученеие рекуррентной сети прогнозирования временного ряда, с двумя рекуррентными слоями и с использованием прореживания.</h3>

```python
model = Sequential()
model.add(GRU(32,
    dropout = 0.1,
    recurrent_dropout = 0.5,
    return_sequences = True,
    input_shape = (None, AEW.shape[-1]))) 
model.add(GRU(32, activation = 'relu', dropout = 0.1,
    recurrent_dropout = 0.5))
model.add(Dense(1))
model.summary()
```

```python
model.compile(optimizer=RMSprop(), loss='mae')
history = model.fit_generator(train_gen,                 
                             steps_per_epoch=500,    
                             epochs=10,                            
                             validation_data=val_gen,       
                             validation_steps=val_steps)
```

```python
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(1, len(loss) + 1)
plt.figure()
plt.grid(True)
plt.plot(epochs, loss, 'bo', label='Ошибка на обучающей выборке')
plt.plot(epochs, val_loss, 'b', label='Ошибка на проверочной выборке')
plt.title('Ошибка на учебной и проверочной выбоорках свёрточной нейронной сети')
plt.legend()
plt.show()
```

```python
mean_mae = np.mean(history.history['loss'])
print(mean_mae)
```

```python
mean_mae = np.mean(history.history['val_loss'])
print(mean_mae)
```
