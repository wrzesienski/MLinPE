```python
import numpy as np
import math
import matplotlib.pyplot as plt
import pandas as pd
```

```python
N = 64 # точек за период
I1 = 7 # амплитуда основной частоты
I5 = 20/5 # амплитуда k-й частоты
fi1 = 120 # фаза основ чатсоты
fi2 = 120 # фаза k-й частоты
F = 50 # частота
w = 2*math.pi*F 
k = 5
t=0
deltaT = 1/F/N
size = 10 * 64 

def signal_tec(F, size, Imax, noise):
    signal = []
    t=0
    w = 2*math.pi*F
    for i in range(size):
        It = Imax * math.sin(w*t + math.radians(fi1))
        signal.append(It)
        t+=deltaT
    return signal


def signal_next(F, size, Imax, noise):
    signal2 = [signal[-1]]
    t = 0
    w = 2*math.pi*F
    for i in range(size):
        It2 = Imax * math.sin(w*t + math.radians(fi1))
        signal2.append(It2)
        t+= deltaT
    return signal2

def signal_after_next(F, size, Imax, noise):
    signal3 = [signal2[-1]]
    t=0
    w = 2*math.pi*F
    for i in range(size):
        It = Imax * math.sin(w*t + math.radians(fi1))
        signal3.append(It)
        t+=deltaT
    return signal3

```

Фаза А

```python
# Делаем КЗ
images = []
while len(images) <= 1000 :
    F = np.random.randint(45,55, dtype=int)
    size = 1000
    Imax = np.random.randint(50,51, dtype=int)
    noise = np.random.randint(1,5, dtype=int)
    
    Fkz = np.random.randint(45,55, dtype=int)
    Imaxkz = np.random.randint(10,40, dtype=int)
    noisekz = np.random.randint(5,7, dtype=int)
    
    signal = signal_tec(F, 1200, Imax, noise)
    signal_kz = signal_next(Fkz, 1200, Imaxkz, noisekz)
    signal2 = signal_after_next(F, 600, Imax, noise)
    
    image1 = signal + signal_kz + signal2
    images.append(image1)
```
