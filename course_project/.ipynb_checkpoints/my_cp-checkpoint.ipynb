### Функция нормализации

```python
def normalize(data):
    mean = data[['Ia','Ib','Ic']].mean(axis=0)
    data[['Ia','Ib','Ic']] -= mean
    std = data[['Ia','Ib','Ic']].std(axis=0)
    data[['Ia','Ib','Ic']] /= std
    
    mean = data[['Ua','Ub','Uc']].mean(axis=0)
    data[['Ua','Ub','Uc']] -= mean
    std = data[['Ua','Ub','Uc']].std(axis=0)
    data[['Ua','Ub','Uc']] /= std
    
    return data
```

### Функция - генератор, формирующая окна из массива данных

```python
def generatorXY(X, Y, lookback, delay, min_index, max_index, 
             batch_size=128, step=1):
    number = -1
    while 1:
        number+=1
#         Если номер выбираемой осциллограммы превышает количество осциллограмм,
#        то начинаем отсчет заново
        if(number > Y.shape[0] - 1):
            number = 0
            
        osc = X[number]
        answer = Y[number]

        max_index = len(osc) - delay - 1

        i = min_index + lookback
        
        if i + batch_size >= max_index:
            rows = np.arange(i, min(i + batch_size, max_index))
        rows = np.arange(i, min(i + batch_size, max_index))  
        i += len(rows)
        samples = np.zeros((len(rows), 
                               lookback//step,
                               5))
        targets = np.zeros((len(rows)))

        for j, row in enumerate(rows):
            indices = range(rows[j] - lookback, rows[j], step)
            samples[j] = osc[indices]
            
            targets[j] = answer[j]

        print(targets)
        yield samples, answer
```

```python
# Количество выборок в прошлом
lookback = 2800
# Количество выборок в будущем 
delay = 0
# Шаг генератора
step = 1
# КОличество окон в пакете данных
batch_size = 1


train_gen = generatorXY(X_train, 
                        y_train,
                        lookback=lookback,
                        delay=delay,
                        min_index=0,
                        max_index=len(X_train[0]),
                        batch_size=batch_size,
                        step=step)

val_gen = generatorXY(X_val,
                      y_val,
                      lookback=lookback,
                      delay=delay,
                      min_index=0,
                     max_index=len(X_val[0]),
                     batch_size=batch_size,
                     step=step)

test_gen = generatorXY(X_test,
                      y_test,
                      lookback=lookback,
                      delay=delay,
                      min_index=0,
                      max_index=len(X_test[0]),
                      batch_size=batch_size,
                      step=step)
```

```python
NB_CLASSES = 5
INPUT_SHAPE = (lookback // step, 3)
INPUT_SHAPE
```

### Функция преобразования бинарного ряда в сответствующий класс

```python
def predict_classes(pred, true):
    
    preds = []
    trues = []
    for i in range(len(pred)):
        if pred[i][0] == 1.0:
            preds.append(0)
        elif pred[i][1] == 1.0:
            preds.append(1)
        else:
            preds.append(2)
        
            
    for i in range(len(true)):
        if true[i][0] == 1.0:
            trues.append(0)
        elif true[i][1] == 1.0:
            trues.append(1)
        else:
            trues.append(2)
            
    return preds, trues
        
```

### Функция преобразования чисел в прогнозе в бинарный ряд

```python
def binary_series(pred):
    for i in range(len(pred)):
        for j in range(len(pred[i])):
            if pred[i][j] >= 0.5:
                pred[i][j] = 1.0
            else:
                pred[i][j] = 0.0
    return pred
```

### Функция для матрицы ошибок

```python
def confus_matrix(pred, true):
    confusion_matrix2 = confusion_matrix(pred, true)
    cm_df = pd.DataFrame(data = confusion_matrix2, columns = ['0','1','2'], index = ['0','1','2'])
    
    f, ax = plt.subplots(figsize=(10,8))
    ax = sns.heatmap(cm_df, annot=True, fmt='d')
    return ax
```

```python
confus_matrix(true_class, pred_class)
```

### Функция для расчета метрик качества

```python
def quality_metrics(y_true, y_pred):
    precision_arr = []
    recall_arr = []
    f1_arr = []

    for step in range(1397):
        precision = precision_score(y_true[step], y_pred[step])
        recall = recall_score(y_true[step], y_pred[step])
        f1 = f1_score(y_true[step], y_pred[step])

        precision_arr.append(precision)
        recall_arr.append(recall)
        f1_arr.append(f1)
        
    return np.mean(precision_arr), np.mean(recall_arr), np.mean(f1_arr)
```

```python
prec, rec, f1 = quality_metrics(y_true, y_pred)
```
